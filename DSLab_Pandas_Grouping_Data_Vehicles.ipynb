### Did cars consume more gas in the eighties?

```python
Car_copy['City MPG'] = Car_copy['City MPG'] * 0.425 
Car_copy.rename(columns={'City MPG': 'City_KPL'}, inplace=True)
Car_copy['Highway MPG'] = Car_copy['Highway MPG'] * 0.425 
Car_copy.rename(columns={'Highway MPG': 'Highway_KPL'}, inplace=True)
Car_copy
```

```python
CityKPL_avg.info()
```

```python
sns.barplot(data= CityKPL_avg, x="Year", y="City_KPL")
```

```python
def decades(x):
    if x >1980 and x <= 1989:
        return "80s"
    elif x> 1989 and x<= 1999:
        return "90s"
    elif x> 1999 and x<= 2010:
        return"10s"
    else:
        
        return "20s"
        
```

```python
Car_copy["Decade"] = cars["Year"].apply(lambda x: decades(x))
```

```python
Car_copy
```

```python
A=Car_copy.groupby(["Decade", "Make"])["CO2 Emission Grams/Km"].mean()
DF= pd.DataFrame(A)
DF
```

```python
B=DF.sort_values(by='CO2 Emission Grams/Km', ascending=False)
df=pd.DataFrame(B)
df
```

```python
filtered_df = df[df['CO2 Emission Grams/Km'].notnull()]
filtered_df
```

Which brands are more environment friendly?

```python
# Fisker,smart,Daihatsu,Fiat most environment friendly brands
filtered_df["CO2 Emission Grams/Km"].min()
```

Did cars consume more gas in the eighties?

```python
filtered_df = filtered_df.groupby('Decade')['CO2 Emission Grams/Km'].mean().reset_index()
filtered_df
```

```python
#Did cars consume more gas in the eighties-Ans:Yes,80s car had more emissions
sns.barplot(data=filtered_df, x="Decade", y="CO2 Emission Grams/Km")
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
TRANS_COPY = Car_copy
```

```python
def check_transmission(Transmission):
    if Transmission.startswith ("Auto"):
        return "Auto"
    else:
        return "Manual"
    
```

```python
TRANS_COPY["check_transmission"] = TRANS_COPY.Transmission.apply(check_transmission)
TRANS_COPY
```

```python
Trans = TRANS_COPY.groupby('check_transmission')['Fuel Barrels/Year'].mean().reset_index()
Trans
```

```python
# Do cars with automatic transmission consume more fuel than cars with manual transmission? Yes,automatic cars consume more fuel
```

```python
plt.figure(figsize=(4,4))

sns.barplot(data=Trans,x="check_transmission", y="Fuel Barrels/Year")
```

Use `groupby` and `agg` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
TRANS_COPY.groupby('check_transmission').agg({'City_KPL':'mean', 'Transmission': 'count'})
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
TRANS_COPY.groupby('check_transmission').agg({'City_KPL':'min'})
```
