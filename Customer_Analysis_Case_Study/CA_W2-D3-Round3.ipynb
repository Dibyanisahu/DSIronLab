### STEP11: Plotting pearsons coefficient

```python
sns.set_theme(style="white")
corr = Numerical.corr()
mask = np.triu(Numerical.corr())
f, ax = plt.subplots(figsize=(10, 10))
cmap = sns.color_palette("Blues")

sns.heatmap(corr, 
            mask=mask, 
            cmap=cmap, 
            vmax=.3, 
            center=0,
            square=True, 
            linewidths=3, 
            cbar_kws={"shrink": .5}
           )
```

```python
def sns_hist_plot(df):
    colors=["coral","red","chocolate","bisque","gold","yellowgreen",
            "lawngreen","lightgreen","darkgreen","cyan","dodgerblue",
            "royalblue","mediumpurple","darkviolet","magenta","deeppink","pink"]
    col_num=len(list(df.columns))
    if col_num<4:
        fig1, axes = plt.subplots(1,col_num,figsize=(5*col_num,6))
    else:
        fig1, axes = plt.subplots(int(np.ceil(col_num/4)),4,figsize=(16,1.2*col_num))
    sns.set(style="darkgrid")
    for i,column in enumerate(df.columns):
        if col_num<=4:
            sns.histplot(x=column, kde=True, data=df, ax=axes[0,i],color=colors[i])
        else:
            sns.histplot(x=column, kde=True, data=df, ax=axes[int(np.ceil(i/4))-1,i%4],color=colors[i])
    return fig1
```

```python
i=sns_hist_plot(Numerical)
```
