```python
def RenameState(ST):
    if ST == "Cali":
        return "California"
    elif ST == "AZ":
        return "Arizona"
    elif ST == "WA":
        return "Washington"
    elif ST == np.nan:
        pass
    else:
        return ST
    
```

```python
CA_df['ST'] = CA_df['ST'].apply(RenameState)
```

```python
CA_df
```

```python
print(CA_df['ST'].unique())
```

```python
def cleanGender(GENDER):
    if GENDER == "Femal":
        return "F"
    elif GENDER == "Male":
        return "M"
    elif GENDER == "female":
        return "F"
    elif GENDER == np.nan:
        pass
    else:
        return GENDER
```

```python
CA_df['GENDER'] = CA_df['GENDER'].apply(cleanGender)
```
